```python
# Paso 2: VISUALIZACIÓN DE LOS DATOS
# 2.1: Importaremos librerías necesarias para generar gráficos que ayuden a entender mejor los datos como las relaciones entre varaibles y su distribución
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2.2 realizamos los histogramas para la categoría de variables continuas
# utilizamos sns.histplot para crear los histogramas, kde=True añade una estimación de densidad de kernel para ayudar a ver la distribución de la variable
# utilizamos color para color de histograma, bins=30 define numero de barras, plt.grid añade cuadrícula al gráfico para facilitar visualización.
continuous_variables = ['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
%matplotlib inline
for cont_var in continuous_variables:
   plt.figure(figsize=(10, 6))
   sns.histplot(data=df, x=cont_var, kde=True, color='blue', bins=30)
plt.title(f"Histograma de '{cont_var}'")
plt.xlabel(cont_var)
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
plt.clf()
```

```python
# 2.3 Visualización de variables categóricas con barras.
#donde plt.figure configura tamaño del gráfico, plt.title agrega título a cada variable, el x e y label establce las etqiea

categorical_variables = ['male', 'education', 'currentSmoker', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'TenYearCHD']
for cat_var in categorical_variables:
  plt.figure(figsize=(10,6))
  sns.countplot(data=df, x=cat_var, color="skyblue")
  plt.title(f"Frecuencia de {cat_var}")
  plt.xlabel(cat_var)
  plt.ylabel('Frecuecia')
  plt.grid(True, axis='y')
  plt.show()
```

```python
# 2.4 Construimos matriz de correlacion entre las variables continuas
# nos permitirá evaluar como las variables continuas están rlaciondas entre sí. Los valoers oscilan entre -1 a 1.
# 1 indica correlación positiva, -1 correlación negativa, 0 no relación.
#si hubiera variables altamente erlacioandas pueden causar problemas de colinearidad
# corr_matrix... calcular correlación de pearson
corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
#heatmap crea el mapa de calor con las especificaciones, annot true son valoers de las celdas y fmt=2f formaeta los números on dos decimales
# cmao coolwarm utiliza paleta de colores del azul (correlación negativa) a rojo (correlación positiva)
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python
# podemos vel en la matriz de correlación que las variables que se correlacionan positivamente son las más importanetes
# entre ellas: current smoker con cigsperday (0.77), la diaBP con sisBP(0.78), ambas con prevalentHyp (0.7 y0.62 respectivamsnet), tambiénglucosa con diabetes(0.62)
#todos los hallazgos son esperables.
```

```python
#Paso 3. Imputacion y normalización de variables.
# 3.1: Detectar los valores faltantes ó valores nulos.
#df.isnull...cuenta los valoers nulos en cada columnadel data frmae
print("Valores nulos por columna")
print(df.isnull().sum())
```

```python
#3.2 Imputación de valores faltantes: reemplaza valoes faltantescon estimaciones razonables.
#para variables continuas usaremos la media
for col in continuous_variables:
  df[col].fillna(df[col].mean(), inplace=True)
#para variables categóricas usaremos la moda
for col in categorical_variables:
  df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Ahora verificamos que ya no hay valores nulos
print("\nDespués de la imputación:")
print(df.isnull().sum())
```

```python
#3.3: Normalización de los datos:
#proceso de escalar caracteristicas a una escala similar. entonces sean comparables. para que la media sea de cero y la DE de 1.
from sklearn.preprocessing import StandardScaler
# Noralizaremos las variables continuas
scaler = StandardScaler()
df[continuous_variables] = scaler.fit_transform(df[continuous_variables])
#Visualizaremos las características después de la normalización
print(df[continuous_variables].describe())
```
